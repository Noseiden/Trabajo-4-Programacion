**Atributos**  

**date:** La fecha y hora en que se registraron los datos.   
**Appliances:** El consumo de energía en vatios por hora de los electrodomésticos en el hogar.   
**lights:** El consumo de energía en vatios por hora de las luces en el hogar.    
**T1, T2, T3, T4, T5, T6, T7, T8, T9:** Temperaturas en grados Celsius de diferentes ubicaciones dentro de la casa (por ejemplo, T1 podría ser la temperatura en una habitación, T2 en otra, y así sucesivamente).  
**RH_1, RH_2, RH_3, RH_4, RH_5, RH_6, RH_7, RH_8, RH_9:** Humedad relativa en porcentaje de diferentes ubicaciones dentro de la casa.  
**T_out:** Temperatura exterior en grados Celsius.  
**Press_mm_hg:** Presión atmosférica en milímetros de mercurio.  
**RH_out:** Humedad relativa exterior en porcentaje.  
**Windspeed:** Velocidad del viento en metros por segundo.  
**Visibility:** Visibilidad en kilómetros.  
**Tdewpoint:** Punto de rocío en grados Celsius.  
**rv1, rv2:** Variables aleatorias generadas uniformemente.
